# Caminho para os dados de teste e treino

```python
teste =userdata.get('path_test_data')
treino = userdata.get('path_train_data')
```

# Funções importantes

```python
def print_heatmap(data,title=""):
    heatmap = sns.heatmap(data)
    heatmap.set_title(title)
    heatmap.set(xlabel='', ylabel='')
    print(heatmap)
```

```python
def jet_image(caminho):
  with h5py.File(caminho,'r') as arquivo:
    data = arquivo.get("jetImage")
    dataset_jet_image = np.array(data)
  return dataset_jet_image
```

# Entendendo os dados

## Dados de teste

```python
teste_dataset = jet_image(teste)
```

```python
id = 3
x_shape =teste_dataset[id].astype('uint8').shape[0]
y_shape = teste_dataset[id].astype('uint8').shape[1]
reshape_dataset = teste_dataset[id].astype('uint8').reshape((x_shape,y_shape))

print_heatmap(reshape_dataset,'Teste')
```

## Dados de treino

```python
treino_dataset = jet_image(treino)
```

```python
id = 3
x_shape =treino_dataset[id].astype('uint8').shape[0]
y_shape = treino_dataset[id].astype('uint8').shape[1]
reshape_dataset = treino_dataset[id].astype('uint8').reshape((x_shape,y_shape))

print_heatmap(reshape_dataset,'Treino')
```
